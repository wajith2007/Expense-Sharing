```python
import numpy as np
from prettytable import PrettyTable

friends=[]
balances=None
history=[]
def add_users():
    global balances
    n=int(input("Number of users: "))
    for _ in range(n):
        name=input("Enter name: ").lower()
        friends.append(name)
    balances=np.zeros(len(friends))
def equal_expense():
    payer=input("Payer: ").lower()
    amount=float(input("Amount: "))
    desc=input("Description: ").lower()
    people=input("Participants : ").split(",")
    people=[p.strip() for p in people]
    share=amount/len(people)
    payer_idx=friends.index(payer)
    for p in people:
        idx=friends.index(p)
        if p!=payer:
            balances[idx]-=share
            balances[payer_idx]+=share
    history.append((payer,amount,desc))
def custom_expense():
    payer=input("Payer: ").lower()
    desc=input("Description: ").lower()
    payer_idx=friends.index(payer)
    total=0
    print("Enter amount for each user:")
    for i,p in enumerate(friends):
        amt=float(input(p+": "))
        total+=amt
        if p!=payer:
            balances[i]-=amt
            balances[payer_idx]+=amt
    history.append((payer,total,desc))
def show_balances():
    t=PrettyTable(["Person","Balance"])
    for i,p in enumerate(friends):
        t.add_row([p,round(balances[i],2)])
    print(t)
def settle():
    debtors=[]
    creditors=[]
    for i,amt in enumerate(balances):
        if amt<0:debtors.append([friends[i],-amt])
        elif amt>0:creditors.append([friends[i],amt])
    i=j=0
    print("\nSettlements:")
    while i<len(debtors) and j<len(creditors):
        d,da=debtors[i]
        c,ca=creditors[j]
        pay=min(da,ca)
        print(d,"pays",c,"₹",round(pay,2))
        debtors[i][1]-=pay
        creditors[j][1]-=pay
        if debtors[i][1]==0:i+=1
        if creditors[j][1]==0:j+=1
def menu():
    add_users()
    while True:
        print("\n1.Equal\n2.Custom\n3.Balances\n4.Settle\n5.Exit")
        ch=input("Choice: ")
        if ch=="1":equal_expense()
        elif ch=="2":custom_expense()
        elif ch=="3":show_balances()
        elif ch=="4":settle()
        elif ch=="5":break
        else:print("Invalid")

menu()
```
